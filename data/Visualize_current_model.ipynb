```python
import requests
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from template_data import sparce_template, dense_template


URL = 'http://127.0.0.1:8000/'


def post_data(x):
    rx = requests.post(URL + 'post', json={'x': x[0], 'y': x[1]})
    return rx.json()


def get_all_data():
    rx = requests.get(URL + 'all_data')
    # print(x, x.text, type(x.text))
    # print(x.json())
    x = []
    y = []
    id = []
    for i in rx.json():
        x.append(i['x'])
        y.append(i['y'])
        id.append(i['cluster_id'])
    return np.array([x,y,id]).T
```

```python
X = get_all_data()
```

```python
plt.figure(figsize =(8, 8))
plt.scatter(X.T[0], X.T[1], c = X.T[2], cmap ='rainbow')
plt.show()
```
